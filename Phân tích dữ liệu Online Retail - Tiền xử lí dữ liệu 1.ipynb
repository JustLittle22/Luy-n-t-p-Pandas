```python
import pandas as pd
import numpy as np
data = pd.read_csv('OnlineRetail.csv')
data
```

```python
data.info()
```

```python
data['month'] = pd.to_datetime(data['InvoiceDate'], format='%m/%d/%Y %H:%M').dt.month
data
```

```python
data.loc[(data['month'] <= 12), 'Previous'] = 4
data.loc[(data['month'] <= 9), 'Previous'] = 3
data.loc[(data['month'] <= 6), 'Previous'] = 2
data.loc[(data['month'] <= 3), 'Previous'] = 1
data
```

```python
data['Previous'] = data['Previous'].astype(int)
data
```

```python
data['Amount'] = data['Quantity'] * data['UnitPrice']
data
```

```python
conditions = [(data['Country'] == 'United Kingdom') & (data['Previous'] == 4),
             (data['Country'] == 'France') & (data['Previous'] == 4)]
choices = [0.1, 0.05]
data['Discount rate'] = np.select(conditions, choices, default=0)
data
```

```python
data['Total'] = round(data['Amount'] - data['Amount'] * data['Discount rate'],2)
data
```
